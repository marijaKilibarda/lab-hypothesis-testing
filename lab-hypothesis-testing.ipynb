```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here
#the hypothesis, an independent two-sample t-test

#H0: dragon_hp = grass_hp, there is no difference in average HP between types Dragon and Grass 
#H1: dragon_hp > grass_hp, Pokemons of type Dragon have a higher average HP

#significance level = 5%

dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
grass_hp = df[df['Type 1'] > 'Grass']['HP']


st.ttest_ind(dragon_hp,grass_hp, alternative='greater', equal_var=False)
```

```python
# Because p_value is lower than significance level, 
# p-value (0.0063) is much smaller than the significance level of 0.05 
# the null hypothesis is rejected. There is strong statistical evidence 
# at the 5% significance level to conclude that Dragon-type PokÃ©mon have, 
# on average, higher HP than Grass-type PokÃ©mon. 
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#the hypothesis, CLT stats
#H0: mean_legendary == mean_non_legendary, the mean stat for Legendary Pokemons equals that of Non-Legendary Pokemon
#H1: mean_legendary != mean_non_legendary,the means differ 

mean_legendary = df[df['Legendary'] == True]['HP'].mean()
mean_non_legendary = df[df['Legendary'] == False]['HP'].mean()

var_hp_legendary = df[df['Legendary'] == True]['HP'].var(ddof=1)
var_hp_non_legendary = df[df['Legendary'] == False]['HP'].var(ddof=1)

n_legendary = df[df['Legendary'] == True]['HP'].count()
n_non_legendary = df[df['Legendary'] == False]['HP'].count()

# Step 2: Calculate the Standard Error (SE)
SE_hp = np.sqrt((var_hp_legendary / n_legendary) + (var_hp_non_legendary / n_non_legendary))

# Step 3: Calculate the z-score (as the sample size is relatively large, we can use z-test)
z_hp = (mean_legendary - mean_non_legendary) / SE_hp

# Step 4: Calculate the p-value (two-tailed test, large sample size -> normal approximation)
p_value_hp = st.t.sf(abs(z_hp), n_legendary-1)*2 

z_hp, p_value_hp
```
